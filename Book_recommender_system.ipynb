# Popularity based recommender system

```python
ratings_that_in_books= ratings.merge(books, on='ISBN')
```

```python
ratings.shape, ratings_that_in_books.shape
```

```python
print(ratings_that_in_books.head(100).to_string())
```

```python
ratings_top_df= ratings_that_in_books.groupby('Book-Title').count()['Book-Rating'].reset_index()
ratings_top_df.rename(columns={'Book-Rating':'num_ratings'}, inplace=True)
ratings_top_df
```

```python
ratings_top_df.shape
```

```python
ratings_avg_df= ratings_that_in_books.groupby('Book-Title')['Book-Rating'].mean().reset_index()
ratings_avg_df.rename(columns={'Book-Rating':'avg_rating'}, inplace=True)
ratings_avg_df
```

```python
popularity_rating_df= ratings_top_df.merge(ratings_avg_df, on='Book-Title')
popularity_rating_df.head()
```

```python
popularity_rating_books= popularity_rating_df[popularity_rating_df['num_ratings']>=250].sort_values('avg_rating', ascending=False).head(50)
```

```python
popularity_rating_books
```

```python
popular_books=popularity_rating_books.merge(books, on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_rating']]
```

```python
popular_books.reset_index(inplace=True)
popular_books
```

```python
popular_books.drop('index', inplace=True, axis=1)
```

```python
popular_books
```

```python
popular_books.to_csv('popular_50_books.csv', index=False)
```

# Collaborative Filtering Based Recommender for books

---



```python
df=ratings.merge(books, on='ISBN')
df
```

```python
pade_likhe_users= df.groupby('User-ID').count()['Book-Rating']>200
pade_likhe_users= pade_likhe_users[pade_likhe_users].index
pade_likhe_users
```

```python
df= df[df['User-ID'].isin(pade_likhe_users)]
```

```python
famous_books= df.groupby('Book-Title').count()['Book-Rating']>=50
famous_books= famous_books[famous_books].index
```

```python
df= df[df['Book-Title'].isin(famous_books)]
```

```python
df
```

```python
pivot_table=df.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
```

```python
pivot_table.fillna(0, inplace=True)
pivot_table
```

```python
#import cosine similarity
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pivot_table)
```

```python
similarity_score
```

```python
similarity_score.shape
```

```python
np.save("similarity_score.npy",similarity_score)
```

```python
def recommend(book_name):
  index= np.where(pivot_table.index==book_name)[0][0]
  similar_items= sorted(list(enumerate(similarity_score[index])), key=lambda x:x[1], reverse=True)[1:6]
  data=[]
  for i in similar_items:
    item= []
    temp_df= books[books['Book-Title']==pivot_table.index[i[0]]]
    item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
    item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
    item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
    data.append(item)
  return data


```

```python
print(recommend('1984'))
```

```python
print(recommend('The Da Vinci Code'))
```
